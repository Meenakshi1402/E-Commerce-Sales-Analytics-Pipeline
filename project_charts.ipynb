```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the data
monthly = pd.read_csv("reports/monthly_revenue.csv", parse_dates=["sale_date"])
category = pd.read_csv("reports/category_summary.csv")

# 2. Plot Monthly Revenue Trend
plt.figure()
plt.plot(monthly["sale_date"], monthly["revenue"])
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Plot Category Breakdown
plt.figure()
plt.bar(category["category"], category["total"])
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
